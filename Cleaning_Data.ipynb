```python
import pandas as pd
import numpy as np
df = pd.read_csv("raw data/aac_shelter_outcomes.csv")

df.head()
```

```python
dfdogs = df[df["animal_type"] == "Dog"]

dfdogs.head(30)
```

```python
dfdogs_cleaned = dfdogs.drop(columns = ["age_upon_outcome", "animal_type", "color", "name", "outcome_subtype", "monthyear"])


dfdogs_cleaned.head(15)
```

```python
dfdogs_cleaned["date_of_birth"] = pd.to_datetime(dfdogs_cleaned["date_of_birth"])
dfdogs_cleaned["datetime"] = pd.to_datetime(dfdogs_cleaned["datetime"])
dfdogs_cleaned.head()
```

```python
breedsframe =dfdogs_cleaned["breed"].value_counts().to_frame()
breedslist = breedsframe.index
dog_df_breeds = dfdogs_cleaned.loc[dfdogs_cleaned["breed"].isin(breedslist[0:10])]

dog_df_breeds
```

```python
dog_df_breeds["age"]  = (dog_df_breeds["datetime"] - dog_df_breeds["date_of_birth"])/ np.timedelta64(1, 'Y')

dog_df_breeds

dog_df_breeds["Success"] = 0
dog_df_breeds["Sex"] = ""

dog_df_breeds.loc[dog_df_breeds.outcome_type == "Adoption", "Success" ] = 1

dog_df_breeds.loc[dog_df_breeds.sex_upon_outcome == "Spayed Female", "Sex" ] = "Female"
dog_df_breeds.loc[dog_df_breeds.sex_upon_outcome == "Intact Female", "Sex" ] = "Female"
dog_df_breeds.loc[dog_df_breeds.sex_upon_outcome == "Neutered Male", "Sex" ] = "Male"
dog_df_breeds.loc[dog_df_breeds.sex_upon_outcome == "Intact Male", "Sex" ] = "Male"

dog_df_breeds = dog_df_breeds.drop(columns= ["outcome_type", "sex_upon_outcome", "date_of_birth", "datetime"])
dog_df_breeds = dog_df_breeds.dropna()

dog_df_breeds = dog_df_breeds[dog_df_breeds.Sex != ""]

dog_df_breeds.head()
```

```python
breed_df = pd.read_csv("raw data/Breed_data.csv")

breed_df.head()
```

```python
breed_df = breed_df.rename(columns={'Breed': 'breed'})

breed_df.head()
```

```python
shelter_df = dog_df_breeds.merge(breed_df, how='inner', on='breed', right_index=False)

shelter_df["Percent through Life"] =  shelter_df["age"]/ shelter_df["life_expectancy"]

shelter_df.to_csv("Shelter Data.csv")

shelter_df = shelter_df.drop(columns=["animal_id", "age", "life_expectancy"])
shelter_df
```

```python
# Generate our categorical variable lists
breed_cat = shelter_df.dtypes[shelter_df.dtypes == "object"].index.tolist()
shelter_df[breed_cat].nunique()
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import StandardScaler,OneHotEncoder
import tensorflow as tf


# Create a OneHotEncoder instance
enc = OneHotEncoder(sparse=False)

# Fit and transform the OneHotEncoder using the categorical variable list
encode_df = pd.DataFrame(enc.fit_transform(shelter_df[breed_cat]))

# Add the encoded variable names to the dataframe
encode_df.columns = enc.get_feature_names(breed_cat)
encode_df.head()
```

```python
shelter_df = shelter_df.merge(encode_df,left_index=True, right_index=True)
shelter_df = shelter_df.drop(breed_cat,1)
shelter_df.head()
```

```python
# Split our preprocessed data into our features and target arrays
y = shelter_df["Success"].values
X = shelter_df.drop(["Success"],1).values

# Split the preprocessed data into a training and testing dataset
X_train, X_test, y_train, y_test = train_test_split(X, y, random_state=10)
```

```python
# Create a StandardScaler instances
scaler = StandardScaler()

# Fit the StandardScaler
X_scaler = scaler.fit(X_train)

# Scale the data
X_train_scaled = X_scaler.transform(X_train)
X_test_scaled = X_scaler.transform(X_test)
```

```python
# Define the model - deep neural net, i.e., the number of input features and hidden nodes for each layer.
number_input_features = len(X_train[0])
hidden_nodes_layer1 =  15
hidden_nodes_layer2 = 6

nn = tf.keras.models.Sequential()

# First hidden layer
nn.add(
    tf.keras.layers.Dense(units=hidden_nodes_layer1, input_dim=number_input_features, activation="relu")
)


# Second hidden layer
nn.add(tf.keras.layers.Dense(units=hidden_nodes_layer2, activation="relu"))

# Output layer
nn.add(tf.keras.layers.Dense(units=1, activation="sigmoid"))
# Check the structure of the model
nn.summary()
```

```python
# Compile the model
from tensorflow.keras.callbacks import ModelCheckpoint
nn.compile(loss="binary_crossentropy", optimizer="adam", metrics=["accuracy"])
```

```python
# Train the model
fit_model = nn.fit(X_train,y_train,epochs=200) 
```

```python
# Evaluate the model using the test data
model_loss, model_accuracy = nn.evaluate(X_test_scaled,y_test,verbose=2)
print(f"Loss: {model_loss}, Accuracy: {model_accuracy}")
```
